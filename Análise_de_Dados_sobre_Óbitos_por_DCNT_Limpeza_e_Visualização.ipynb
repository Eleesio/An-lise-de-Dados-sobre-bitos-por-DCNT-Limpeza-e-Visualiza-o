# **Análise de Dados sobre Óbitos por DCNT: Limpeza e Visualização**

***Neste exercício, trabalhamos com um conjunto de dados sobre óbitos por Doenças Crônicas Não Transmissíveis (DCNT) para realizar uma análise exploratória. O objetivo principal foi preparar os dados para análise e criar visualizações que nos permitissem entender melhor a distribuição das idades dos óbitos, assim como os fatores como sexo, raça/cor, escolaridade e categorias CID. Através da limpeza e visualização dos dados, buscamos identificar padrões e gerar insights útei***

Primeiramente, carregamos o conjunto de dados usando a biblioteca pandas e tratamos alguns problemas comuns de qualidade dos dados. Especificamente:

  Leitura dos dados: Utilizamos o método read_csv para carregar o arquivo CSV com o separador correto (;), garantindo que erros em linhas malformadas fossem ignorados. A codificação do arquivo foi configurada para utf-8 para evitar problemas com caracteres especiais.

  Tratamento das datas: A coluna de datas foi convertida para o formato datetime, utilizando o pd.to_datetime(). Durante essa conversão, lidamos com erros em casos de datas inválidas, além de garantir que as datas fossem interpretadas no formato correto, com o dia vindo antes do mês.

  Cálculo e correção de idades: A coluna de idades foi corrigida para garantir que valores inválidos fossem descartados e os valores nulos fossem preenchidos com base na diferença entre a data de nascimento e a data de óbito. Utilizamos a função apply() para corrigir valores negativos ou absurdos nas idades.

  Substituição de valores nulos: Para as colunas que representavam sexo, raça/cor, escolaridade e categoria CID, substituímos valores nulos por 'Desconhecido', garantindo que essas variáveis não tivessem lacunas.

  Visualizações: Criamos uma série de gráficos utilizando o seaborn para ilustrar diferentes aspectos dos dados:

  Distribuição das idades dos óbitos: A primeira visualização foi um histograma com estimativa de densidade, mostrando como as idades estavam distribuídas entre os óbitos.

  Boxplot por sexo: Criamos um boxplot para comparar a distribuição das idades dos óbitos entre os sexos.

  Contagem de óbitos por raça/cor e escolaridade: Utilizamos gráficos de barras para visualizar a distribuição dos óbitos em função dessas duas variáveis.

  Contagem de óbitos por categoria CID: Exibimos a contagem de óbitos por categorias CID, agrupando categorias menos frequentes em "Outros" para evitar excesso de rótulos.
        
  Idade por sexo e raça/cor: Finalmente, fizemos um boxplot horizontal, segmentado por sexo e raça/cor, para examinar como a idade variava entre essas duas dimensões.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar o dataset com o separador correto
url = 'dados_cronicas_ses_2023.csv'  # Substitua com o caminho correto do arquivo
df = pd.read_csv(url, sep=';', on_bad_lines='skip', encoding='utf-8')

# Exibir as primeiras linhas para revisão
print(df.head())

# Corrigir a conversão das colunas de data
df['dt_nascimento'] = pd.to_datetime(df['dt_nascimento'], errors='coerce', dayfirst=True)
df['dt_obito'] = pd.to_datetime(df['dt_obito'], errors='coerce', dayfirst=True)

# Calcular a idade corretamente, levando em consideração possíveis erros na coluna 'nu_idade'
df['nu_idade'] = pd.to_numeric(df['nu_idade'], errors='coerce')  # Converter para numérico, ignorando erros

# Recalcular a idade com base na diferença de datas, se necessário
df['nu_idade'] = df['nu_idade'].fillna(((df['dt_obito'] - df['dt_nascimento']).dt.days // 365).astype(float))

# Corrigir valores inválidos na idade (por exemplo, idades negativas ou absurdas)
df['nu_idade'] = df['nu_idade'].apply(lambda x: x if x > 0 else None)

# Verificar se há valores nulos em 'nu_idade' após a correção
print(f"Valores nulos em 'nu_idade': {df['nu_idade'].isnull().sum()}")

# Substituir valores nulos nas colunas importantes (exemplo: sexo, raça/cor, escolaridade)
df['sg_sexo'].fillna('Desconhecido', inplace=True)
df['tp_raca_cor'].fillna('Desconhecido', inplace=True)
df['tp_escolaridade'].fillna('Desconhecido', inplace=True)
df['categoria_cid_causa_basica'].fillna('Desconhecido', inplace=True)

# Exibir algumas estatísticas descritivas da coluna de idade
print(df['nu_idade'].describe())

# Melhorar o estilo das visualizações
sns.set(style="whitegrid", palette="muted", font_scale=1.2)

# Visualização 1: Distribuição de Idades dos Óbitos
plt.figure(figsize=(14, 8))
sns.histplot(df['nu_idade'], kde=True, color='cornflowerblue', bins=30)
plt.title('Distribuição de Idades dos Óbitos por DCNT', fontsize=18, fontweight='bold')
plt.xlabel('Idade', fontsize=16)
plt.ylabel('Frequência', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Visualização 2: Boxplot de Idades por Sexo
plt.figure(figsize=(14, 8))
sns.boxplot(x='sg_sexo', y='nu_idade', data=df, palette='pastel')
plt.title('Distribuição de Idade dos Óbitos por Sexo', fontsize=18, fontweight='bold')
plt.xlabel('Sexo', fontsize=16)
plt.ylabel('Idade', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Visualização 3: Contagem de Óbitos por Raça/Cor
plt.figure(figsize=(14, 8))
sns.countplot(x='tp_raca_cor', data=df, palette='coolwarm', order=df['tp_raca_cor'].value_counts().index)
plt.title('Contagem de Óbitos por Raça/Cor', fontsize=18, fontweight='bold')
plt.xlabel('Raça/Cor', fontsize=16)
plt.ylabel('Contagem de Óbitos', fontsize=16)
plt.xticks(rotation=45, fontsize=14)
plt.yticks(fontsize=14)
plt.tight_layout()
plt.show()

# Visualização 4: Contagem de Óbitos por Escolaridade
plt.figure(figsize=(14, 8))
sns.countplot(x='tp_escolaridade', data=df, palette='Set2', order=df['tp_escolaridade'].value_counts().index)
plt.title('Contagem de Óbitos por Escolaridade', fontsize=18, fontweight='bold')
plt.xlabel('Escolaridade', fontsize=16)
plt.ylabel('Contagem de Óbitos', fontsize=16)
plt.xticks(rotation=45, fontsize=14)
plt.yticks(fontsize=14)
plt.tight_layout()
plt.show()

# Visualização 5: Contagem de Óbitos por Categoria CID
# Agrupar categorias menos frequentes em "Outros" para evitar excesso de rótulos
threshold = 50
top_categories = df['categoria_cid_causa_basica'].value_counts()
df['categoria_cid_causa_basica'] = df['categoria_cid_causa_basica'].apply(
    lambda x: x if top_categories[x] > threshold else 'Outros'
)

plt.figure(figsize=(14, 8))
sns.countplot(
    x='categoria_cid_causa_basica',
    data=df,
    palette='viridis',
    order=df['categoria_cid_causa_basica'].value_counts().index
)
plt.title('Contagem de Óbitos por Categoria CID', fontsize=18, fontweight='bold')
plt.xlabel('Categoria CID', fontsize=16)
plt.ylabel('Contagem de Óbitos', fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=14)
plt.tight_layout()
plt.show()

# Visualização 6: Idade dos Óbitos por Sexo e Raça/Cor (gráfico horizontal para clareza)
plt.figure(figsize=(14, 12))
sns.boxplot(
    x='nu_idade',
    y='sg_sexo',
    hue='tp_raca_cor',
    data=df,
    palette='coolwarm'
)
plt.title('Distribuição de Idade dos Óbitos por Sexo e Raça/Cor', fontsize=18, fontweight='bold')
plt.xlabel('Idade', fontsize=16)
plt.ylabel('Sexo', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(title='Raça/Cor', fontsize=14, title_fontsize=16)
plt.tight_layout()
plt.show()
```

# Através da limpeza dos dados e das visualizações realizadas, conseguimos obter uma visão clara da distribuição de óbitos por DCNT, segmentada por várias características demográficas. As correções na idade e nos dados ausentes permitiram uma análise mais precisa, e os gráficos gerados forneceram insights valiosos sobre os padrões de óbitos por sexo, raça/cor, escolaridade e categorias CID.
